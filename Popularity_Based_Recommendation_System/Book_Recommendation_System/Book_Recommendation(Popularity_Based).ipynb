```python
books.dropna(inplace=True)
books.isnull().sum()
```

```python
ratings.dropna(inplace=True)
ratings.isnull().sum()
```

```python
rating = ratings.merge(books, on ='ISBN')
```

```python
num_rating=rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df=num_rating.rename(columns={'Book-Rating':'num_ratings'})
num_rating_df
```

```python
avg_rating = rating.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the 'Book-Rating' column to 'avg_ratings'
avg_rating_df = avg_rating.rename(columns={'Book-Rating': 'avg_ratings'})

# Display the resulting DataFrame
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popularity = popular_df.merge(books, on='Book-Title')
```

```python
Popular_books = popularity[['Book-Title', 'num_ratings', 'avg_ratings', 'Book-Author', 'Image-URL-M']]
```

```python
popular_books=Popular_books.drop_duplicates('Book-Title')
```
